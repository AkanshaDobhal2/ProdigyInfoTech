```python
survived = df[df['survived']==1]
not_survived = df[df['survived']==0]
total_passenger=len(df['survived'])
print("not_survided:",len(not_survived))
print("survived:",len(survived))
print("Total number of survived columns:",len(df['survived']))

print('The percentage of survived was: {:.2f}%'.format((len(survived)/total_passenger)*100))
print('The percentage of not survived was:{:.2f}%'.format((len(not_survived)/total_passenger)*100))
```

```python

y= [len(survived['survived']),len(not_survived['survived'])]
x = ['Survived','Not_survived']
bars=plt.bar(x,y,color='green')

# giving title to the plot
plt.title('Count of surviveds on Titanic')
     
    # giving X and Y labels
plt.xlabel("How many survived")
plt.ylabel("Count")
     
for b in bars:
    height= b.get_height()
    plt.annotate(f'{height}' ,xy=(b.get_x() + b.get_width() /2,height),xytext=(0,3), 
                 textcoords="offset points", ha='center', va='bottom')
plt.show()
```

```python
gender_survivors = df[df['survived'] == 1].groupby('sex').size()
gender_non_survivors = df[df['survived'] == 0].groupby('sex').size()
x = np.arange(len(gender_survivors))
```

```python
bar_width = 0.35

plt.bar(x - bar_width/2, gender_survivors, bar_width, label='Survivors', color='orange')
plt.bar(x + bar_width/2, gender_non_survivors, bar_width, label='Non-Survivors', color='blue')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Survivors vs. Non-Survivors by Gender')
plt.xticks(x, gender_survivors.index)
plt.legend()
plt.show()
```
