10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# Create the new column 'total_bedroom_size' based on the criteria
#continuous variable to a categorical variable.
df['total_bedroom_size'] = pd.cut(df['total_bedrooms'],
                                  bins=[0, 10, 1000, df['total_bedrooms'].max()], labels=['small', 'medium', 'large'])

# Display the first 5 rows of the dataframe with the new column
df.head()
```
